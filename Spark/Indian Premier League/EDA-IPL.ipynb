```python
#total number of teams participating (Team names)
team1_names = ipl_matches_df.team1.unique()
team1_names.tolist()
#print("total number of team :",team1_names)
team2_names = ipl_matches_df.team2.unique()
team2_names.tolist
#print("total number of team :",team2_names)
total_teams = list(set(team1_names) & set(team2_names))
print(total_teams)
print("total team :",len(total_teams))
```

```python
#Details on Toss won by each team, Total Matches played so far, total matches being won list.
Total_match = pd.DataFrame({'Total Matches played': ipl_matches_df['team1'].value_counts() + ipl_matches_df['team2'].value_counts(),
                           'Total won': ipl_matches_df['winner'].value_counts(),
                           'Toss won': ipl_matches_df['toss_winner'].value_counts(), 
                          'Total lost': ((ipl_matches_df['team1'].value_counts() + ipl_matches_df['team2'].value_counts())
                                         - ipl_matches_df['winner'].value_counts())})

Total_match=Total_match.reset_index()
Total_match.rename(columns = {'index':'Teams'}, inplace = True)
Total_match
```

```python
 #Maximum Toss Won:
plt.subplots(figsize=(10,7))
ax=ipl_matches_df['toss_winner'].value_counts().plot.barh(width=0.8)
plt.title("Maximum Toss Won")
```

```python
#Teams who had won Toss and Won the match
Tosswin_matchwin=ipl_matches_df[ipl_matches_df['toss_winner']==ipl_matches_df['winner']]
labels=['Yes','No']
slices=[len(Tosswin_matchwin),(636-len(Tosswin_matchwin))]
plt.pie(slices,labels=labels,startangle=90,shadow=True,explode=(0,0),autopct='%1.1f%%',colors=['g','r'])
plt.title("Teams who had won Toss and Won the match")
```

```python
#Plotting the above data Analysis - Total Won
Total_won = ipl_matches_df.winner.value_counts()
ax = Total_won.plot(kind='bar', title ="No. of wins by each team - Overall",figsize=(10,7),legend=True, fontsize=12)
ax.set_xlabel("Teams",fontsize=12)
ax.set_ylabel("count",fontsize= 12)
```

# calculate a Team's winning percentage

```python
Win_per=Total_match['Total won']/Total_match['Total Matches played']
win_per=pd.DataFrame({"win_percentage":Win_per})
win_per=win_per.round(2)
Total_match=Total_match.join(win_per)
Total_match
```

```python
plt.subplots(figsize=(10,7))
sns.barplot('Total won', 'Teams', hue='win_percentage', data=Total_match, palette='Paired');
```

```python
batsmen_summary = ball_matches_df.groupby('batsman').agg({'ball': 'count','batsman_runs': 'sum'})
batsmen_summary.rename(columns={'ball':'balls', 'batsman_runs': 'runs'}, inplace=True)
batsmen_summary = batsmen_summary.sort_values(['balls','runs'], ascending=False)[:20]
batsmen_summary.head(10)
```

# How is the strike rate of batsmen calculated?

# Strike Rate = (Total Runs Scored)/(Total Balls Faced)*100

```python
batsmen_summary['batting_strike_rate'] = batsmen_summary['runs']/batsmen_summary['balls'] * 100
batsmen_summary['batting_strike_rate'] = batsmen_summary['batting_strike_rate'].round(2)
batsmen_summary.head(10)
```

```python
plt.subplots(figsize=(10,7))
sns.barplot('runs', 'balls', data=batsmen_summary, palette='Paired',hue='batting_strike_rate');
```

```python
venue_ser = ipl_matches_df['venue'].value_counts()
venue_df = pd.DataFrame(columns=['venue', 'matches'])
for items in venue_ser.iteritems():
    temp_df = pd.DataFrame({
        'venue':[items[0]],
        'matches':[items[1]]
    })
    venue_df = venue_df.append(temp_df, ignore_index=True)
```

```python
plt.subplots(figsize=(10,6))
plt.title("IPL Venues")
sns.barplot(x='matches', y='venue', data=venue_df);
```

```python
team_wins_ser = ipl_matches_df['winner'].value_counts()

team_wins_df = pd.DataFrame(columns=["team", "wins"])
for items in team_wins_ser.iteritems():
    temp_df1 = pd.DataFrame({
        'team':[items[0]],
        'wins':[items[1]]
    })
    team_wins_df = team_wins_df.append(temp_df1, ignore_index=True)
```

# Most Valuable Player

```python
mvp_ser = ipl_matches_df['player_of_match'].value_counts()

mvp_ten_df = pd.DataFrame(columns=["player", "wins"])
count = 0
for items in mvp_ser.iteritems():
    if count>9:
        break
    else:
        temp_df2 = pd.DataFrame({
            'player':[items[0]],
            'wins':[items[1]]
        })
        mvp_ten_df = mvp_ten_df.append(temp_df2, ignore_index=True)
        count += 1    
```

```python
plt.subplots(figsize=(10,6))
plt.title("Top Ten IPL Players")
sns.barplot(x='wins', y='player', data=mvp_ten_df, palette='Paired');
```

```python
#Season's Winner
season_winner = ipl_matches_df.drop_duplicates(subset=['season'], keep='last')[['season', 'winner']].reset_index(drop=True)
season_winner
```

```python
#Number of Wickets taken:
bowler_wickets = ball_matches_df.groupby('bowler').aggregate({'ball': 'count', 'total_runs': 'sum', 'player_dismissed' : 'count'})
bowler_wickets.columns = ['runs','balls','wickets']
TopBowlerBywickets = bowler_wickets.sort_values(['wickets'], ascending=False)[:20]
TopBowlerBywickets
```

```python
plt.subplots(figsize=(10,6))
plt.title("Top Ten IPL Bowler Players")
sns.barplot(x='wickets', y='runs', data=TopBowlerBywickets.head(15), palette='Paired');
```
